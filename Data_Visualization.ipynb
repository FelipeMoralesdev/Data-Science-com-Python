##01. Conhecendo a biblioteca Matplotlib


```python
import pandas as pd
```

```python
url = "https://raw.githubusercontent.com/alura-cursos/bibliotecas_visualizacao/main/Dados/imigrantes_canada.csv"
```

```python
df = pd.read_csv(url)
```

```python
df.set_index('País', inplace = True)
```

```python
df
```

```python
anos = list(map(str, range(1980, 2014)))
```

```python
anos
```

```python
brasil = df.loc[ 'Brasil', anos]
```

```python
brasil
```

```python
brasil_dict = {'Ano': brasil.index.tolist(), 'Imigrantes': brasil.values.tolist()}
```

```python
brasil_dict
```

```python
dados_brasil = pd.DataFrame(brasil_dict)
```

```python
dados_brasil
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(16,4))
plt.plot(dados_brasil['Ano'], dados_brasil['Imigrantes'])
plt.title('Imigração do Brasil para o Canadá')
plt.xlabel('Ano')
plt.ylabel('Número de imigrantes')
plt.xticks(['1980', '1985', '1990', '1995', '2000', '2005', '2010'])
plt.show()
```

###Desafio

```python
argentina = df.loc[ 'Argentina', anos]
argentina_dict = {'Ano': argentina.index.tolist(), 'Imigrantes': argentina.values.tolist()}
dados_argentina = pd.DataFrame(argentina_dict)
```

```python
plt.plot(dados_brasil['Ano'], dados_brasil['Imigrantes'], label='Brasil')
plt.plot(dados_argentina['Ano'], dados_argentina['Imigrantes'], label='Argentina')
plt.title('Imigração do Brasil e Argentina para o Canadá')
plt.xlabel('Ano')
plt.ylabel('Número de imigrantes')
plt.xticks(['1980', '1985', '1990', '1995', '2000', '2005', '2010'])
plt.legend()
plt.show()
```

##02. Criando figuras com Matplotlib

```python
fig, ax = plt.subplots(figsize=(8,4))
ax.plot(dados_brasil['Ano'], dados_brasil['Imigrantes'])
ax.set_title('Imigração brasileira para Canadá\n1980 - 2013')
ax.set_xlabel('Ano')
ax.set_ylabel('Número de imigrantes')
ax.xaxis.set_major_locator(plt.MultipleLocator(5))
plt.show()
```

```python
fig, axs = plt.subplots(1,2, figsize=(15,5))
axs[0].plot(dados_brasil['Ano'], dados_brasil['Imigrantes'])
axs[0].set_title('Imigração brasileira para Canadá\n1980 - 2013')
axs[0].set_xlabel('Ano')
axs[0].set_ylabel('Número de imigrantes')
axs[0].xaxis.set_major_locator(plt.MultipleLocator(5))
axs[0].grid()

axs[1].boxplot(dados_brasil['Imigrantes'])
axs[1].set_title('Boxplot da imigração brasileira para Canadá\n1980 - 2013')
axs[1].set_xlabel('Brasil')
axs[1].set_ylabel('Imigrantes')
axs[1].grid()
```

```python
fig, axs = plt.subplots(2,2, figsize=(10,6))
fig.subplots_adjust(hspace=0.5, wspace=0.3)
fig.suptitle('Comparação entre países')

axs[0,0].plot(df.loc['Brasil', anos])
axs[0,0].set_title('Brasil')

axs[0,1].plot(df.loc['Colômbia', anos])
axs[0,1].set_title('Colômbia')

axs[1,0].plot(df.loc['Argentina', anos])
axs[1,0].set_title('Argentina')

axs[1,1].plot(df.loc['Peru', anos])
axs[1,1].set_title('Peru')

for ax in axs.flat:
  ax.xaxis.set_major_locator(plt.MultipleLocator(5))
  ax.set_xlabel('Ano')
  ax.set_ylabel('Número de imigrantes')

ymin = 0
ymax= 7000

for ax in axs.ravel():
  ax.set_ylim(ymin, ymax)

plt.show()
```

##03. Customizando com Matplotlib

```python
fig, ax = plt.subplots(figsize=(8,4))
#Espessura da linha do gráfico (lw) e marcadores (marker)
ax.plot(dados_brasil['Ano'], dados_brasil['Imigrantes'], lw=3, marker='o')
#Tamanho do título (fontsize) e posição (loc)='left'
ax.set_title('Imigração do Brasil para o Canadá\n1988 a 2013', fontsize=18, loc='left')
ax.set_xlabel('Ano', fontsize=14)
ax.set_ylabel('Número de imigrantes', fontsize=14)
ax.xaxis.set_tick_params(labelsize=12)
ax.yaxis.set_tick_params(labelsize=12)
ax.xaxis.set_major_locator(plt.MultipleLocator(5))
#Grade e estilo da linha (linestyle)
plt.grid(linestyle='--')
plt.show()
```

```python
# Criar a figura e os subplots
fig, axs = plt.subplots(2, 2, figsize=(14, 8))

# Ajustar os espaçamentos entre os subplots
plt.subplots_adjust(wspace=0.3, hspace=0.4)

# Adicionando um título geral para os subplots
fig.suptitle('Vendas no período de janeiro a dezembro de 2022 nas lojas A,B,C e D', fontsize=20)

# Lista de cores
cores_lojas = ['darkviolet', 'green', 'darkblue', 'coral']

#Loop para plotar e customizar os subplots
for i, ax in enumerate(axs.flat):
    ax.plot(dados_lojas.loc[dados_lojas.index[i]], color=cores_lojas[i], lw=3)
    ax.set_title(f'Vendas na loja {dados_lojas.index[i]}', loc='left', fontsize=16)
    ax.set_xlabel('Mês', fontsize=14)
    ax.set_ylabel('Número de vendas', fontsize=14)
    ax.tick_params(labelsize=12)
    ax.grid(color='lightgrey')
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)

for ax in axs.ravel():
  ax.set_ylim(0, 500)
```
